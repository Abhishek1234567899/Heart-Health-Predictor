```python
# Importing necessary libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Loading the dataset
df = pd.read_csv("heart.csv")

# Basic Information
print("Missing Values:", df.isnull().sum().sum())
print(df.info())

# Summary Statistics
print("Mean:\n", df.mean())
print("Median:\n", df.median())
print("Mode:\n", df.mode().iloc[0])  # .iloc[0] to avoid all modes
print("Skewness:\n", df.skew())

# Correlation Matrix Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()

# Target Distribution
sns.histplot(df["target"], kde=True)
plt.title('Target Variable Distribution')
plt.show()

# Sex vs Target (Bar Plot)
pd.crosstab(df.sex, df.target).plot(kind="bar", figsize=(10, 6), color=['blue', '#AA1111'])
plt.title('Heart Disease Frequency for Sex')
plt.xlabel('Sex (0 = Female, 1 = Male)')
plt.xticks(rotation=0)
plt.legend(["Haven't Disease", "Have Disease"])
plt.ylabel('Frequency')
plt.show()

# Age Distribution with KDE
plt.figure(figsize=(10, 6))
sns.histplot(df['age'], kde=True, bins=20, color="purple")
plt.title('Age Distribution')
plt.xlabel('Age')
plt.show()

# Cholesterol vs Target (Boxplot)
plt.figure(figsize=(10, 6))
sns.boxplot(x='target', y='chol', data=df, palette='Set2')
plt.title('Cholesterol Levels by Heart Disease')
plt.xlabel('Heart Disease (0 = No, 1 = Yes)')
plt.ylabel('Cholesterol')
plt.show()

# Chest Pain Type Distribution
plt.figure(figsize=(8, 6))
sns.countplot(x='cp', data=df, palette='husl')
plt.title('Chest Pain Type Distribution')
plt.xlabel('Chest Pain Type (0-3)')
plt.ylabel('Frequency')
plt.show()

# Max Heart Rate Achieved Distribution by Target
plt.figure(figsize=(10, 6))
sns.kdeplot(data=df, x='thalach', hue='target', fill=True, common_norm=False, palette="crest", alpha=0.5)
plt.title('Max Heart Rate Achieved by Heart Disease')
plt.xlabel('Max Heart Rate (thalach)')
plt.ylabel('Density')
plt.show()


# Age vs Maximum Heart Rate Achieved
plt.figure(figsize=(10, 6))
sns.scatterplot(x='age', y='thalach', hue='target', palette='coolwarm', data=df)
plt.title('Age vs Maximum Heart Rate Achieved')
plt.xlabel('Age')
plt.ylabel('Max Heart Rate Achieved')
plt.legend(title="Heart Disease (0 = No, 1 = Yes)")
plt.show()

# Frequency of Fasting Blood Sugar > 120
sns.countplot(x='fbs', data=df, palette='cool')
plt.title('Frequency of Fasting Blood Sugar > 120 mg/dl')
plt.xlabel('FBS (0 = <120 mg/dl, 1 = >120 mg/dl)')
plt.ylabel('Count')
plt.show()

# Slope of Peak Exercise ST Segment by Target
pd.crosstab(df['slope'], df['target']).plot(kind='bar', figsize=(10, 6), color=['#4CAF50', '#FF5722'])
plt.title('Slope of Peak Exercise ST Segment vs Heart Disease')
plt.xlabel('Slope of Peak Exercise ST Segment (0-2)')
plt.ylabel('Frequency')
plt.legend(["No Disease", "Disease"])
plt.show()

# Distribution of Exercise-Induced Angina
sns.countplot(x='exang', data=df, palette='viridis')
plt.title('Exercise-Induced Angina Distribution')
plt.xlabel('Exercise-Induced Angina (0 = No, 1 = Yes)')
plt.ylabel('Frequency')
plt.show()
```
